# 2.2 정확도 측정

- RMSE 평균 제곱근 오차 : Root mean square error )
- 위꺾쇠가 붙으면 햇이라고 한다. 
- 오차 : y (실제값) - y^ 와이햇(= 예측값) 
- 모델이 예측한 값을 실제값에다 빼주는것. 
- 모델이 예측한 값 (= 추정값) 

```python
import numpy as np

## RMSE

## 파이썹에서 곱하기 2개는 ** 두제곱을 의미,  *** 세제곱
def RMSE(y_true, y_pred):
    residualzancha = np.array(y_true) - np.array(y_pred)
    return np.sqrt(np.mean((  residualzancha )**2))
```

```python
#정확도 계산하는 rmse 다시만들기 
def RMSE(y_true, y_pred):
    return np.sqrt(np.mean((np.array(y_true) - np.array(y_pred))**2))
```

```python
'''
모델별 rmse 계산 함수
모델별 예측치의 정확도를 계산 
'''
def score(model):
    # 모델이 아직 없지. 모델이 되는 함수를  나중에  정의한 뒤에 인수로 넣어 주자. def 0000 
    id_pairs =zip(x_test['user_id'], x_test['movie_id'])
    y_pred = np.array([model(user,movie) for (user,movie) in id_pairs])# 리스트를 만들어주기
    y_true = np.array(x_test['rating'])
    return RMSE(y_true, y_pred)
```
